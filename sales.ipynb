# SALES OF DIFFERENT CHHANEL

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load the dataset into df
df = pd.read_csv("advertising.csv")

# Step 3: Check the first few rows to confirm structure
print(df.head())
```

# Plot()
#sum()

```python
# Sum total spend per channel
budget_totals = df[["TV", "Radio", "Newspaper"]].sum()

# Plot the total budget as bar chart
budget_totals.plot(kind='bar', title="Total Advertising Spend by Channel")
plt.ylabel("Total Budget (in $1000s)")
plt.show()
```

#group()

```python
# Create a TV budget group (bucket every $50)
df["TV_Budget_Group"] = (df["TV"] // 50) * 50

# Group by TV budget range and sum sales
tv_sales = df.groupby("TV_Budget_Group")["Sales"].sum()

tv_sales.plot(kind="line", marker='o', title="Sales by TV Budget Group")
plt.xlabel("TV Budget Group")
plt.ylabel("Total Sales")
plt.show()
```

```python
sns.scatterplot(data=df, x="TV", y="Sales")
plt.title("TV Advertising vs Sales")
plt.xlabel("TV Budget ($1000s)")
plt.ylabel("Sales (Units)")
plt.show()
```

```python
sns.scatterplot(data=df, x="Radio", y="Sales", color="green")
plt.title("Radio Advertising vs Sales")
plt.xlabel("Radio Budget")
plt.ylabel("Sales")
plt.show()
```

```python
sns.pairplot(df)
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.hist(df["Sales"], bins=10, color='skyblue', edgecolor='black')
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
# Create a new column: High vs Low TV spend
df["TV_Spend_Level"] = ["High" if val > 150 else "Low" for val in df["TV"]]

sns.boxplot(x="TV_Spend_Level", y="Sales", data=df)
plt.title("Sales by TV Spend Level")
plt.xlabel("TV Budget Level")
plt.ylabel("Sales")
plt.show()
```
